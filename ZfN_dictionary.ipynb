```python
import glob
import codecs
import regex
```

## Tokens extrahieren und bereinigen

### Liste mit allen Zeilen erstellen

```python
filelist = glob.glob("data/*/*/*.txt")
lines = []

for f in filelist:
    datei = codecs.open(f, "r", encoding="utf8")
    
    for line in datei:
        lines.append(line)
    datei.close()

print("Elemente in Zeilenliste: ", len(lines))
print("Element 0: ", lines[0])
print("Element 39: ", lines[39])
```

### Liste mit allen Tokens erstellen

**Es wird stumpf am Leerzeichen gesplittet, da nltk-Tokenizer Punkte entfernt, die auf Abkürzungen verweisen können.**

```python
#an Leerzeichen splitten
tokens = []

for line in lines:
    tokens.extend(line.split(" "))

print("Anzahl Wörter (mit anhängender Punctuation): ", len(tokens))
print("Token aus Zeilenliste[0]: ", tokens[0:9])
```

#### Überflüssige Punctuation(-Elemente) entfernen
##### Alleinstehende Punctuation-Elemente

```python
#https://realpython.com/python-enumerate/

counter = 0
counter_del = 0
del_list = []   #wenn Element direkt gelöscht wird, ändern sich Indices aller 
                #anderen Elemente, deshalb Indices zuerst in separater Liste speichern

for i, item in enumerate(tokens):
    if regex.findall(r'^(\p{P}+)$', item):  #alle Strings, die aus einem oder mehreren Punctuation-Zeichen bestehen
        counter = counter+1
        del_list.append(i) 
        counter_del = counter_del+1
print("Gefundene Elemente: \t", counter)
print("Elemente in del_list: \t", counter_del)

#Indices von groß nach klein sortieren, damit sich beim Iterieren nicht die
#Position der restlichen Listenelemente ändert
for index in sorted(del_list, reverse=True):
    print("Gelöschtes Element: \t", tokens.pop(index))

print("Nachher (noch ", len(tokens), "Elemente): ", tokens)
```

##### Punctuation aus Wörtern entfernen

Alles außer Bindestrichen (Double Oblique Hyphen, Soft Hypen oder Hyphen-Minus), Apostrophen und Punkten entfernen.
--> Abkürzungen bleiben erkennbar

```python
print("Vorher: ", tokens)
for i, item in enumerate(tokens):
    tokens[i] = regex.sub(r'[\(\)\[\]\{\}=§%#<>\*"%:,;„“‚‘\!\?]', "", item)  #einfacher, mit inklusivem Character Set zu arbeiten, als etwas mit Ausnahmen zu konstruieren

print("Nacher: ", tokens)
```

### Begriffe für Wörterbuch sammeln

**Schritt 1:** Token-Liste Tokens in Types umwandeln --> Mehrfachelemente werden entfernt --> Liste wird schlanker

#### Umwandlung von Tokens in Types

```python
types = list(set(tokens)) 
print("Anzahl der Typen: ", len(types))
```

#### Teilliste numbers

**Anmerkung:** Die Liste wird nicht in das Wörter-Dictionary integriert, sondern  kann zur Erstellung eines number-Dictionarys genutzt werden. 

```python
numbers = []

for item in types:
    if regex.findall(r'(\d+[a-zA-ZäöüÄÖÜ]+\.?)|(\d+ſ\w+)', item) or regex.findall(r'(\d+\p{P}+)', item): #beliebig lange Zahl gefolgt von Buchstaben mit optionalem Punkt oder beliebig lange Zahl gefolgt von Lang-s mit nachfolgenden Wortzeichen ODER beliebig lange Zahl gefolgt von Punctuation
        numbers.append(item)

print("Anzahl Elemente in numbers:", len(numbers))
print(numbers)
```

##### DAWG-Datei für numbers anlegen

```python
numbers_dawg = []

#Ziffern aus Listenelementen entfernen
for item in numbers:
    if regex.findall(r'\d', item):
        item_oZ = regex.sub(r'\d', " ", item)                    #item_oZ = item ohne Ziffer, Ziffern entfernen und durch nichts ersetzen
        numbers_dawg.append(item_oZ)

print("Anzahl Elemente in numbers_dawg:", len(numbers_dawg))
print(numbers_dawg)

numbers_dawg = sorted(list(set(numbers_dawg)))
print("Anzahl Elemente in aufgeräumtem numbers_dawg:", len(numbers_dawg))

#numbers-Patterns in Datei schreiben
with open("numbers.dawg.txt", "a") as f:
    for entry in numbers_dawg:
        f.write(entry + '\n')

    print("Numbers-txt erfolgreich angelegt.")

f.close
```

##### Elemente mit Ziffern aus Types-Liste entfernen

```python
numbers_del = []

for i, item in enumerate(types):
    if regex.findall(r'\d+', item): #alles mit Zahl finden
        numbers_del.append(i)

print("Anzahl Elemente in numbers_del", len(numbers_del))

#Indices von groß nach klein sortieren, damit sich beim Iterieren nicht die
#Position der restlichen Listenelemente ändert
for index in sorted(numbers_del, reverse=True):
    print("Gelöschtes Element: \t", types.pop(index))
```

#### Teilliste hyphenated

Wörter mit 2 oder mehr Bindestrichen (inkl. Abkürzungen innerhalb des Wortes)

```python
hyphenated = []

for item in types:
    if regex.findall(r'(\w+(-|⸗|\.-|\.⸗)){1,}\w+\.?', item) or regex.findall(r'(\w+(-|⸗|\.-|\.⸗)){1,}$', item):     #erste Bedingung = ganze Abkürzungen: Zeichenfolge mit Bindestrich oder Punkt und Bindestrich (mindestens einmal), endend mit Zeichenfolge mit oder ohne Punkt
        hyphenated.append(item)                                                                                     #zweite Bedingung = partielle Abkürzungen: String endet mit Bindestrich
        
print(hyphenated)
print("Anzahl Elemente in hyphenated:", len(hyphenated))
```

#### Teilliste abbrevs

##### Abkürzungen/ganze Wörter aus hyphenated-Liste

Einträge im Wörterbuch von deu_latf haben keine Punkte (nur bei Webadressen) --> werden hier deswegen auch weggelassen

```python
abbrevs = []

#Schritt 1: zuerst Wörter trennen
hyph_split = []

for item in hyphenated:
    if '-' in item:
        hyph_split.extend(item.split("-"))
    if '⸗' in item:
        hyph_split.extend(item.split("⸗"))

print("Aufgeteilte Bindestrichwörter:", hyph_split)

#Schritt 2: Aufteilen in "echte" Abkürzungen und ganze Wörter
for item in hyph_split:
    if '.' in item:
        item_oP = item.replace(".", "") #Punkt entfernen --> neue Variable = item ohne Punkt
        abbrevs.append(item_oP)         #entpunktetes Wort wird in abbrevs-Liste aufgenommen
    else:
        dictionary.append(item)         #ganze Wörter aus Abkürzungen werden direkt in die eigentliche dictionary-Liste übertragen

print("abbrevs (", len(abbrevs), "Elemente): ", abbrevs)
print("dictionary (", len(dictionary), "Elemente): ", dictionary)
```

##### Abkürzungen aus types-Liste

```python
#Wörter mit Punkt am Ende
#ja, das nimmt auch ganze Wörter am Satzende mit, aber da der Punkt und später Dopplungen entfernt werden, ist das egal
for item in types:
    if '.' in item and regex.findall(r'^\p{L}|^ſ', item):   #wenn im Item ein Punkt vorhanden ist und mit Buchstabenzeichen (class L = letters) oder Lang-s beginnt:
        item_oP = item.replace(".", "")                     #Punkt entfernen (durch Nichts ersetzen)
        abbrevs.append(item_oP)

print("Neue Anzahl Elemente in abbrevs: ", len(abbrevs))
```

#### Teilliste nouns

Vor allem, um Eigennamen zu erfassen. Erfasst allerdings auch Nicht-Substantive, die am Satzanfang großgeschrieben werden. Ist verkraftbar, da damit zusätzliche Schreibarten erfasst werden

Siehe auch Tanner et. al, Measuring Mass Text Digitization, S. 3: Significant words with capital letter start accuracy

```python
nouns = []  

for item in types:
    if regex.findall(r'^[A-ZÄÖÜ]ſ?\p{L}+$', item): #Ausdrücke, die mit einem Großbuchstaben beginnen und mit einer beliebigen Anzahl Buchstaben enden, sollte alle Ausdrücke mit Punkt oder Bindestrich(en) ausschließen
        nouns.append(item)

print("Anzahl Elemente in nouns:", len(nouns))
```

#### Teilliste most_faulty

Ergibt sich aus den häufig gefundenen Fehlern: Buchstabendreher, Umlaute, diakritische Zeichen

```python
most_faulty = []

#wird viele Wörter finden, die bereits in häufigste drin sind, aber gerade Wörter mit seltener vorkommenden Buchstaben (x, v) werden mit abgedeckt
#h und b sind relativ häufig, aber auch besonders fehleranfällig, ebenso Lang-s
#Buchstabendreher von Großbuchstaben müssen nicht mit aufgenommen werden, da diese bereits durch nouns abgedeckt sind
#Buchstabentauscher der häufigen Buchstaben (a, e, n, u) sind ausgelassen, sonst könnten auch alle token in Wörterbuch übernommen werden

for item in types:
    if regex.search(r'[ſhbpvxy№\u00C0-\u00FF]', item): #bis No. = Buchstabentauscher und seltene Buchstaben, dann Umlaute und diakritische Zeichen als Unicode-Range
        most_faulty.append(item)

print("Anzahl Elemente in most_faulty: ", len(most_faulty))
```

# DICTIONARY

```python
dictionary = hyphenated + abbrevs + nouns + most_common + most_faulty
print("Einträge im Dictionary VOR dem Aufräumen: ", len(dictionary))

#Punkte entfernen
for e, entry in enumerate(dictionary):
    if regex.findall(r'\.(-|⸗)', entry):    #Punkt nicht entfernen, wenn er innerhalb eines Bindestrichwortes steht
        continue
    else:
        dictionary[e] = regex.sub(r'\.', "", entry)  

#leere Strings entfernen
del_empty = []

for i, entry in enumerate(dictionary):
    if entry == '':
        del_empty.append(i)

#Indices von groß nach klein sortieren, damit sich beim Iterieren nicht die
#Position der restlichen Listenelemente ändert
counter = 0
for index in sorted(del_empty, reverse=True):
    dictionary.pop(index)
    counter = counter+1
print("Gelöschte leere Strings:", counter)

#Duplikate entfernen und alphabetisch sortieren
dictionary = sorted(list(set(dictionary)))  

print("Einträge im Dictionary NACH dem Aufräumen: ", len(dictionary))
```

#### Dictionary in txt-Datei schreiben

```python
with open("ZfN_dictionary.txt", "a") as f:
    for entry in dictionary:
        f.write(entry + '\n')

    print("Dictionary-txt erfolgreich angelegt.")

f.close
```
